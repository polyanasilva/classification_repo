## separacao dos dados

```python
x = data.drop(columns=["churn"], axis=1)
y = data["churn"]
x
```

## transformando colunas categoricas em numericas

```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import make_column_transformer

colunas = x.columns

one_hot = make_column_transformer(
    (OneHotEncoder(drop="if_binary"),
     ["pais", "sexo_biologico", "tem_cartao_credito", "membro_ativo"]),
    remainder="passthrough",
    sparse_threshold=0
)

x = one_hot.fit_transform(x)
one_hot.get_feature_names_out(colunas)
```

```python
x = pd.DataFrame(x, columns=one_hot.get_feature_names_out(colunas))
x
```

## transformando variavel alvo em numerica

```python
# Tornando variavel alvo em numerica

label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
y
```

## divisao dos dados

```python
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(
    x, y, test_size=0.2, random_state=5, stratify=y
)
```

## modelo base

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier()
dummy.fit(x_treino, y_treino)

dummy.score(x_teste, y_teste)
```

## arvore de decisao

```python
from sklearn.tree import DecisionTreeClassifier

arvore = DecisionTreeClassifier(random_state=5, max_depth=5)
arvore.fit(x_treino, y_treino)

arvore.score(x_teste, y_teste)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

nome_colunas = ["Alemanha", "Espanha", "França", "Feminino", "tem cartao", 
                "é membro ativo", "score", "idade", "anos_de_cliente", "saldo", 
                "servicos adquiridos", "salario_estimado"]

plt.figure(figsize=(20,10))
plot_tree(arvore, filled=True, feature_names=nome_colunas, class_names=["sim", "nao"], fontsize=7)
plt.show()
```
